**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
x= np.linspace(0,100)
z = x*x
ax1.plot(x,y,color = '#0000CD')
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.plot([20,22],[40,45],color = '#0000CD')
ax2.set_title('zoom')

plt.show()
```
