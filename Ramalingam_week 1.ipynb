```python
import numpy as np
import pandas as pd

import cv2
import matplotlib.pylab as plt

plt.style.use('ggplot')
```

uploading **images**





```python
from google.colab import files
uploaded = files.upload()
```

```python
from PIL import Image
cat_file = next(iter(uploaded))

```

```python
img_mpl = plt.imread(cat_file)
img_cv2 = cv2.imread(cat_file)
img_mpl.shape
```

```python
pd.Series(img_mpl.flatten()).plot(kind='hist',title='Distribution of pixel values')
plt.show()
```

Displaying the images

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.imshow(img_mpl)
ax.axis('off')
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img_mpl[:,:,0], cmap='Reds')
axs[1].imshow(img_mpl[:,:,1], cmap='Greens')
axs[2].imshow(img_mpl[:,:,2], cmap='Blues')
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
axs[0].set_title('Red channel')
axs[1].set_title('Green channel')
axs[2].set_title('Blue channel')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img_cv2)
axs[1].imshow(img_mpl)
axs[0].axis('off')
axs[1].axis('off')
axs[0].set_title('CV2 Image')
axs[1].set_title('Matplotlib Image')
plt.show()
```

```python
img_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots()
ax.imshow(img_cv2_rgb)
ax.axis('off')
plt.show()
```
